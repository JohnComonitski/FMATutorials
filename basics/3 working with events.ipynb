**Get Events by Type and Subtype**

```python
## Get all shots in a match
shots = match.get_events("SHOT")
shots
```

```python
## Get all crosses in a match
crosses = match.get_subtype_events("CROSS")
crosses
```

```python
## Get all successful crosses in a match
# Get Crosses
crosses = match.get_subtype_events("CROSS")
# Filter out crosses that are not passes
crosses = crosses[crosses["Type"] == "PASS"]
crosses
```
