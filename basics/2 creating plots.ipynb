**Creating a Pitch Plot**

The Football pitch is 106 units long by 68 units wide. Starting from the top left corner, the pitch is a plane whose X axis runs from [-53,  53] and y axis runs from [34, -34]. The plot's top left corner coordinates are (-53, 34) while the center is (0, 0).

```python
plot = Plot()
```

**Drawing on the Plot**

Included in the Plot object are several functions that can be called to draw on top of the plot and can be used to annotate events happening on the pitch.

```python
# Draw a point
plot.draw_point(5,5)
```

```python
# Draw a circle
plot.draw_circle(20, 25, 4, c='b')
```

```python
# Draw a box
box = [ [-50, 30], [-45, 30], [-50, 25], [-45, 25] ]
plot.draw_box(box)
```

```python
# Draw a circle
box = []
p1 = (25, 0)
p2 = (0, 0)
plot.draw_line(p1, p2, c='r')
```

```python
# Draw Path
path = [ (-25, 5), (-24, 4), (-23, 3), (-22, 2), (-21, 1), (-20, 0) ]
plot.draw_path(path)
```

```python
# Clear Plot
plot = Plot()
```
